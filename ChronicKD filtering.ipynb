```python
import pandas as pd
```

```python
HF = pd.read_csv('/content/drive/MyDrive/Data Analytics/bq-Distinct HF.csv', sep=',', header = 0)  #Distinc HF has the entire list of patients that are intended to be included in the analysis.
CKD = pd.read_csv('/content/drive/MyDrive/Data Analytics/bq-Distinct CKD.csv', sep=',', header = 0) #This file lists the patients who suffer from some level of Chronic Kidney Disease.

CKD
```

```python
#Compare all of the patients with HF with CKD and find those who suffer from both.
HFCKD = HF
both = []
for row in HFCKD['SUBJECT_ID']:         #Only wish to find the patients that exist in HF dataframe, as it's already filtered.
  if row in (list(CKD['SUBJECT_ID'])):  #If the subject_id exists on both lists then the patient suffers from both conditions.
    both.append(1)
  else:
    both.append(0)
HFCKD['Chronic KD'] = both
HFCKD
```

```python
#Remove the ICD9 code column, as is no longer useful for the analysis.
HFCKD = HFCKD.drop('ICD9_CODE', axis=1)
HFCKD
```

```python
HFCKD.to_csv('/content/drive/MyDrive/Data Analytics/bq-ChronicKD.csv', sep=',', index=False)   #Save the new file with both parameters of data.
```
