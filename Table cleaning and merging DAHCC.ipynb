#ProBNP

```python
#PROBNP
patients = np.unique(BNP_filtered['SUBJECT_ID'])    #Get all unique patients
#patients
table = []    #Resulting values will be stored in a list/array, at the end it's converted into a df.
mean = 0
sd = 0
var = 0
for subject in patients:
  numbers = []
  success = 0
  row = []      #All of the data of each patient is going to be stored in 1 row.
  indices = list(np.where(BNP_filtered['SUBJECT_ID'] == subject)[0])    #Get indices where each patient has entries of data
  try:
    values = BNP_filtered['ProBNP'].iloc[indices]
    #mean = np.mean(BNP_filtered['ProBNP'].iloc[indices].astype(float))  #Calculate the mean value for all rows where a unique patient has data.
    #sd = np.std(BNP_filtered['ProBNP'].iloc[indices].astype(float))     #Same for standard deviation
    mean = np.mean(values.astype(float))        #Calculate the mean value for all rows where a unique patient has data.
    sd = np.std(values.astype(float))           #Same for standard deviation
    row.extend([subject, mean, sd])   #Create the structure of the data that's going to become the final df.
    success += 1
  except:                   #If a non numeric value exists in the actual subject's data.
    for i in values:
      try:                  #Try to convert all of it's data entries to float, if it has more than 1 valid entry calculate parameters and append to table, otherwise eliminate subject from dataframe.
        numbers.append(float(i))
        success += 1
      except:
        None
    if success > 0:
      mean = np.mean(numbers)
      sd = np.std(numbers)
      row.extend([subject, mean, sd])   #Create the structure of the data that's going to become the final df.
  if success > 0:
    table.append(row)

BNP_final = pd.DataFrame(table, columns=['SUBJECT_ID','ProBNP (mean)', 'SD (ProBNP)'])
BNP_final
```
